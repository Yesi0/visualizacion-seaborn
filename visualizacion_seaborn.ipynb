#4.4 Test de hipÃ³tesis

```
sns.lmplot(x='iu', y ='valor_factura', data=tarjetas, col='moroso'); #recordar que col, es para 'abrir' el grafico, o sea separalo, en este caso van haber 2 graficos de morosos, 1 moroso y el otro de no moroso
```

```
from scipy.stats import ranksums #importando libreria para hacer hipotesis
```

Creamos nuestras 2 hipotesis
- Hipotesis null = La distribucion de los grupos moroso y no moroso es la misma
- Hipotesis alterntativa = La distribucion de los grupos moroso y no moroso NO es la misma

```
#Hacemos la consulta con query, query nos permite filtrar bajo condiciones dadas
#con query moroso ==1, EstÃ¡ filtrando el DataFrame para seleccionar solo las filas donde el valor en la columna moroso es igual a 1. Es decir, estÃ¡ seleccionando a los clientes que son morosos.
#El cÃ³digo selecciona los valores de la columna valor_factura del DataFrame tarjetas para aquellos clientes que son morosos (donde el valor en la columna moroso es igual a 1) y almacena el resultado en la variable moroso.
moroso = tarjetas.query("moroso == 1").valor_factura
```

```
moroso
```

Aqui tenemos el total de facturas de las personas  morosas, que son 117

```
no_moroso = tarjetas.query("moroso == 0").valor_factura
```

```
no_moroso
```

Aqui tenemos el total de facturas de las personas NO morosas, que son 1383

```
#con esta funcion podemos ver si la distribucion entre las 2 variables (moroso y no moroso) es significativa
resultado = ranksums(moroso, no_moroso) #usamos ranksums para ver si la distribucion es significativas, y dentro de los parentesis ponemos las variables que vamos a analizar
```

```
resultado
```

Nos enfocamos en el resultado pvalue, que en este caso es un 1%, los valores bajo del 5% sirve para aceptar nuestra hipotesis alternativa (que pusimos al principio)

```
print(f'El p_value de nuestro test es: {resultado.pvalue}')
```

Interpretacion: el pvalue es de 1%, lo que nos permite aceptar uestra hipotesis alternativa, es decir que la distribucion de los morosos y NO morosos no es la misma.
